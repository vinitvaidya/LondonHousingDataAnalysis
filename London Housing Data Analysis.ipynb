### Converting the datatype of "Date" column to "Date-Time" Format

```python
housing_df.head()
```

```python
housing_df.dtypes
```

```python
housing_df['date'] = pd.to_datetime(housing_df['date'])
```

```python
housing_df.dtypes
```

### Adding a new column "Year" in the dataframe, which contains years only

```python
housing_df.head()
```

```python
housing_df['year'] = housing_df['date'].dt.year
```

```python
housing_df.head()
```

### Adding a new column "month" as 2nd column in the dataframe, which contains month only

```python
housing_df.insert(1, 'month', housing_df['date'].dt.month)
```

```python
housing_df.head()
```

### Removing the columns "year" and "month" from the dataframe

```python
housing_df.drop(['month', 'year'], axis = 1, inplace = True)
```

```python
housing_df.head()
```

### Finding maximum and minimum "average price" per year in England

```python
housing_df['year'] = housing_df['date'].dt.year
```

```python
housing_df.head()
```

```python
df1 = housing_df[housing_df.area == 'england']
```

```python
df1
```

```python
#maximum average price
df1.groupby('year').average_price.max()
```

```python
#minimum average price
df1.groupby('year').average_price.min()
```

### Finding maximum and minimum number of crimes recorded per area

```python
housing_df.head()
```

```python
#maximum number of crimes
housing_df.groupby('area').no_of_crimes.max().sort_values(ascending = True)
```

```python
#minimum number of crimes
housing_df.groupby('area').no_of_crimes.min().sort_values(ascending = True)
```

### Displaying total counts of records of each area, where average price is less than 100000

```python
housing_df[housing_df['average_price'] < 100000].area.value_counts()
```
